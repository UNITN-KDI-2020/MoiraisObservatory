```python
import json
import os
import re
import dateparser as dp
```

```python
# Opening JSON file 
with open(os.path.expanduser('~/Desktop/Datasets/Informal_modelling/Data/Meeting_data.json')) as f:
    file = json.load(f)
```

```python
items = []
for line in file:
    title = line['title']
    rating = line['rating']
    dove = str(line['Dove'])
# Split the time into 'From time' & 'To Time' and convert the time into standard format   
    time_lst = (line['Quando'])
    quando = time_lst[0]    
    time = re.search('(?<=Dal)(.*)', quando).group(1)
    quando_list = time.split(' ')
    time_begin = quando_list[1]
    time_end = quando_list[3]
    from_time = str(dp.parse(time_begin))
    to_time = str(dp.parse(time_end))
    
# Calculate the cost from the Prezzo text
    cost_str = str(line['Prezzo'])
    cost = (re.findall('[0-9]+', cost_str))
    if len(cost) == 0:
        Prezzo = 0
    else:
        Prezzo = (cost[0])
    Altre_Informazioni = line['Altre Informazioni']
    materia = line['materia']
    figure = line['figure']
    body = line['body']
 
    item = {'Title': title, 'Rating': rating, 'Dove': dove,'Prezzo': Prezzo, 
          'From_time': from_time, 'To_time': to_time, 
             'Altre Informazioni': Altre_Informazioni, 
             'materia': materia, 'Figure': figure, 'Body': body}
    items.append(item)
with open('Meeting_data_clean.json','w') as f:
    json.dump(items,f,indent=4) 
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
with open('dataset/data_trentino_events.htm', 'r') as f:
    data = f.read().replace('\n', '')
```

```python
soup = BeautifulSoup(data, 'html.parser')
```

```python
events = soup.find_all("div", { "class" : "wrapper" })
```

```python
print("Number of events: ", len(events))
```

```python
for event in events:
    metalist = event.find_all("meta")
    print("img : ", event.find("img").get("src"))
    for meta in metalist:
        print(meta.get('itemprop'), " : ", meta.get('content'))
    
    print("\n==================\n")
```

```python
with open('dataset/data_trentino_cinema.htm', 'r') as f1:
    movies_htm = f1.read().replace('\n', '')
```

```python
movies_htm_parsed = BeautifulSoup(movies_htm, 'html.parser')
```

```python
links = movies_htm_parsed.find_all("a", attrs = { "itemprop": "url" })
links = [link['href'] for link in links]
links = set(links)
```

```python
for uri in links:
    r = requests.get('https://www.crushsite.it' + uri)
    spage = BeautifulSoup(r.text, 'html.parser')
    greybox = spage.find('div', attrs={ 'class' : 'colonnas-1-301ds' })
    print(greybox.find(attrs={ 'itemprop' : 'name' })['content'])
    print(greybox.find(attrs={ 'itemprop' : 'startDate' })['content'])
    print(greybox.find(attrs={ 'itemprop' : 'endDate' })['content'])
    for e in greybox.find_all('span', attrs={ 'class' : 'testo-boxgrigio-grassetto' }):
        print(e.getText())
        print(e.find_next_sibling().getText())
    desc = spage.find('div', attrs={ 'class' : 'colonna-1-5007-testoevi' })
    desc = desc.find('span', attrs={ 'class' : 'testo-boxgrigio-grassetto' })
    if desc is not None:
        s = desc.getText().split('\n\t')
        print(s)
    else:
        print("NO DESCRIPTION!")
    print('\n\n============\n\n')
```

```python
list_htm = requests.get('https://www.visittrentino.info/it/guida/eventi?destination=&category%5B%5D=7863&startdate=&enddate=&submitted=1').text

list_soup = BeautifulSoup(list_htm, 'html.parser')
article_list = list_soup.find_all('article', {'class': 'col col-sm-3 moodboard__item--sm'})
article_links = set([article.find('a', {'role': 'link'})['href'] for article in article_list])
```

```python
for article_link in article_links:
    event_text = requests.get('https://www.visittrentino.info' + article_link).text
    event_soup = BeautifulSoup(event_text, 'html.parser')
    meta = event_soup.find('ul', attrs={ 'class': 'list-unstyled fz24 gallery__highlights__cat' })
    meta_cont = [ el.get_text() for el in meta.find_all('li')]
    print(meta_cont)
    
    header = event_soup.find('div', attrs={ 'class': 'content-block mb60' })
    print("TITLE: ", header.h1.text)
    print("SUBTITLE", header.h2.text)
    
    content_el = event_soup.find('div', attrs={ 'class': 'wysiwyg mt30' })
    content = [p.getText() for p in content_el.find_all('p')]
    print(content)
    sidebar = event_soup.find('div', { 'class': 'col-sm-3 col-sm-push-1 sidebar' })
    bars = sidebar.find_all('div', { 'class': 'mb30' })
    if len(bars) > 1:
        d = bars[1].getText().replace('\n\n\n\n', ':').replace('\n', '').replace('  ', '')
        print(d)
    if len(bars) > 2:
        t = bars[2].getText().replace('\n\n\n\n', ':').replace('\n', '').replace('  ', '')
        print(t)
    org = sidebar.find('address').find('span', {'class': 'strong'})
    if org is not None:
        print(org.getText())
    else:
        print("NO TITLE!")
    print("\n\n===============\n\n")
```

```python
### Extraction of event links for each of which is to be visited
with open('dataset/eventi_ent.htm', 'r') as f1:
    evts_htm = f1.read().replace('\n', '')

eventi = BeautifulSoup(evts_htm, 'html.parser')
table = eventi.find('table', {'id': 'exemple'})
uri_list = [link['href'] for link in table.find_all('a')]
print(uri_list)
```

```python
for uri_string in uri_list:    
    event_text = requests.get(uri_string).text
    event_soup = BeautifulSoup(event_text, 'html.parser')
    meta = event_soup.find('div', attrs={ 'class': 'content-main' })
    t = event_soup.find('div', { 'class' : 'content-title' }).get_text()
    img = event_soup.find('div', { 'class' : 'figure main_image' }).img['src']
    p = meta.find('p', {'class': 'text-justify'})
    if p is not None:
        print("content: ", p.get_text())
    else:
        print("content: No content")
    print("title: ", t)
    print("img_src: ", img)

    identifiers = meta.find_all('div', {'class' : 'col-md-3'})
    for i in range(0, len(identifiers)):
        curr_key = identifiers[i].get_text()
        curr_val = identifiers[i].find_next('div', { 'class': 'col-md-9' }).get_text()
        print(curr_key.replace('\n', '').replace(' ', '_'), ":", curr_val.replace('  ', ''))
        
    print('\n\n================================================\n\n')
```
